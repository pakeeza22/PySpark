```python
from pyspark import SparkContext, SparkConf
import pyspark as ps
spark = ps.sql.SparkSession.builder.master("local").appName("Practice App").getOrCreate() 
```

```python
df = spark.read.format('json').load('trip.json')
print(df)
df.write.format('parquet').bucketBy(100, 'Year', 'Month').mode("overwrite").saveAsTable('table')
```
